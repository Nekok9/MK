```python
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np

# Set the start and end dates for the data
start_date = '1975-01-01'
end_date = '2024-01-01'

# Download the data from FRED using pandas_datareader
japan_gdp = web.DataReader('JPNRGDPEXP', 'fred', start_date, end_date)
france_gdp = web.DataReader('CLVMNACSCAB1GQFR', 'fred', start_date, end_date)

# Convert the data to log scale
log_japan_gdp = np.log(japan_gdp)
log_france_gdp = np.log(france_gdp)

# Apply Hodrick-Prescott filter to extract the cyclical component
japan_cycle, japan_trend = sm.tsa.filters.hpfilter(log_japan_gdp, lamb=1600)
france_cycle, france_trend = sm.tsa.filters.hpfilter(log_france_gdp, lamb=1600)

# Remove NaN values
detrended_data = pd.DataFrame({
    'Japan_Cycle': japan_cycle.squeeze(),
    'France_Cycle': france_cycle.squeeze()
}).dropna()

# Calculate the correlation coefficient
correlation = detrended_data['Japan_Cycle'].corr(detrended_data['France_Cycle'])

# Display the correlation coefficient
print(f"Correlation between detrended Japan and France GDP: {correlation}")

# Plot: Detrended GDP (Cycle) for Japan and France
plt.figure(figsize=(14, 7))
plt.plot(detrended_data['Japan_Cycle'], label="Japan")
plt.plot(detrended_data['France_Cycle'], label="France")
plt.legend()
plt.title("Detrended log(real GDP) 1975Q1-2024Q1")
plt.show()
```
