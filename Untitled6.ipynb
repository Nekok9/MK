```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Tauchenの方法を使ってマルコフ過程を離散化する関数
def tauchen(n, mu, rho, sigma):
    m = 1 / np.sqrt(1 - rho**2)
    state_space = np.linspace(mu - m*sigma, mu + m*sigma, n)
    d = (state_space[n-1] - state_space[0]) / (n-1)
    transition_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if j == 0:
                transition_matrix[i, 0] = norm.cdf((state_space[0] - rho*state_space[i] + d/2)/sigma)
            elif j == n-1:
                transition_matrix[i, n-1] = 1.0 - norm.cdf((state_space[n-1] - rho*state_space[i] - d/2)/sigma)
            else:
                transition_matrix[i, j] = norm.cdf((state_space[j] - rho*state_space[i] + d/2)/sigma) - norm.cdf((state_space[j] - rho*state_space[i] - d/2)/sigma)
    return transition_matrix, state_space

# パラメータを設定する関数
def setPar(sigma=1.5, beta=0.98, rho=0.6, sigma_eps=0.6, a_l=0, a_u=20, NA=401, NH=2, mu_h=-0.7, alpha=0.25, delta=0.03):
    pi, h = tauchen(NH, mu_h, rho, sigma_eps)
    h = np.exp(h)
    param = {'sigma': sigma, 'beta': beta, 'pi': pi, 'h': h, 'NA': NA, 'NH': NH, 'a_l': a_l, 'a_u': a_u, 'alpha': alpha, 'delta': delta}
    return param

# 家計の最適化問題を解く関数
def solve_household(param, r, w):
    NA = param['NA']
    NH = param['NH']
    h = param['h']
    a_l = param['a_l']
    a_u = param['a_u']
    sigma = param['sigma']
    beta = param['beta']
    pi = param['pi']
    a = np.linspace(a_l, a_u, NA)
    util = np.full((NA, NA, NH), -10000.0)
    for ia in range(NA):
        for ih in range(NH):
            for iap in range(NA):
                cons = w*h[ih] + (1.0 + r)*a[ia] - a[iap]
                if cons > 0:
                    util[iap, ia, ih] = cons**(1.0-sigma)/(1.0-sigma)
    v = np.zeros((NA, NH))
    c = np.zeros((NA, NH))
    aplus = np.zeros((NA, NH))
    v_new = np.zeros((NA, NH))
    iaplus = np.full((NA, NH), -10000)
    tol = 1e-6
    test = 10.0
    reward = np.zeros((NA, NA, NH))
    while test > tol:
        for ia in range(NA):
            for ih in range(NH):
                reward[:, ia, ih] = util[:, ia, ih]
                for ihp in range(NH):
                    reward[:, ia, ih] += beta*pi[ih, ihp]*v[:, ihp]
                v_new[ia, ih] = np.max(reward[:, ia, ih])
                iaplus[ia, ih] = np.argmax(reward[:, ia, ih])
        test = np.max(abs(v_new - v))
        v = v_new
    aplus = a[iaplus]
    for ia in range(NA):
        for ih in range(NH):
            c[ia, ih] = w*h[ih] + (1.0 + r)*a[ia] - aplus[ia, ih]
    return aplus, iaplus, c

# 基本ケースのパラメータを設定
param = setPar()
r = 0.04
w = 1
decisions = solve_household(param, r, w)
aplus = decisions[0]
c = decisions[2]
a_l = param['a_l']
a_u = param['a_u']
NA = param['NA']
a = np.linspace(a_l, a_u, NA)

# 貯蓄率を計算してプロット
plt.figure(figsize=(10, 6))
plt.plot(a, aplus[:, 0] / (c[:, 0] + aplus[:, 0]), label='低生産性', color='blue')
plt.plot(a, aplus[:, 1] / (c[:, 1] + aplus[:, 1]), label='高生産性', color='red')
plt.xlabel("現在の資産水準 ($a_t$)")
plt.ylabel("貯蓄率 ($\\frac{a_{t+1}}{c_t + a_{t+1}}$)")
plt.title('異なる生産性状態における現在の資産水準と貯蓄率の関係')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 資本所得税を導入した場合の貯蓄率を計算
def solve_household_with_tax(param, r, w, tau_k):
    NA = param['NA']
    NH = param['NH']
    h = param['h']
    a_l = param['a_l']
    a_u = param['a_u']
    sigma = param['sigma']
    beta = param['beta']
    pi = param['pi']
    a = np.linspace(a_l, a_u, NA)
    util = np.full((NA, NA, NH), -10000.0)
    for ia in range(NA):
        for ih in range(NH):
            for iap in range(NA):
                cons = w*h[ih] + (1.0 + (1 - tau_k) * r)*a[ia] - a[iap]
                if cons > 0:
                    util[iap, ia, ih] = cons**(1.0-sigma)/(1.0-sigma)
    v = np.zeros((NA, NH))
    c = np.zeros((NA, NH))
    aplus = np.zeros((NA, NH))
    v_new = np.zeros((NA, NH))
    iaplus = np.full((NA, NH), -10000)
    tol = 1e-6
    test = 10.0
    reward = np.zeros((NA, NA, NH))
    while test > tol:
        for ia in range(NA):
            for ih in range(NH):
                reward[:, ia, ih] = util[:, ia, ih]
                for ihp in range(NH):
                    reward[:, ia, ih] += beta*pi[ih, ihp]*v[:, ihp]
                v_new[ia, ih] = np.max(reward[:, ia, ih])
                iaplus[ia, ih] = np.argmax(reward[:, ia, ih])
        test = np.max(abs(v_new - v))
        v = v_new
    aplus = a[iaplus]
    for ia in range(NA):
        for ih in range(NH):
            c[ia, ih] = w*h[ih] + (1.0 + (1 - tau_k) * r)*a[ia] - aplus[ia, ih]
    return aplus, iaplus, c

tau_k = 0.3
decisions_with_tax = solve_household_with_tax(param, r, w, tau_k)
aplus_with_tax = decisions_with_tax[0]
c_with_tax = decisions_with_tax[2]

# 貯蓄率を計算してプロット
plt.figure(figsize=(10, 6))
plt.plot(a, aplus[:, 0] / (c[:, 0] + aplus[:, 0]), label='低生産性 - 基本ケース', color='blue')
plt.plot(a, aplus[:, 1] / (c[:, 1] + aplus[:, 1]), label='高生産性 - 基本ケース', color='red')
plt.plot(a, aplus_with_tax[:, 0] / (c_with_tax[:, 0] + aplus_with_tax[:, 0]), label='低生産性 - 資本所得税', linestyle='--', color='blue')
plt.plot(a, aplus_with_tax[:, 1] / (c_with_tax[:, 1] + aplus_with_tax[:, 1]), label='高生産性 - 資本所得税', linestyle='--', color='red')
plt.xlabel("現在の資産水準 ($a_t$)")
plt.ylabel("貯蓄率 ($\\frac{a_{t+1}}{c_t + a_{t+1}}$)")
plt.title('資本所得税導入前後の貯蓄率の比較')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 一括補助金を導入した場合の貯蓄率を計算
def solve_household_with_subsidy(param, r, w, T):
    NA = param['NA']
    NH = param['NH']
    h = param['h']
    a_l = param['a_l']
    a_u = param['a_u']
    sigma = param['sigma']
    beta = param['beta']
    pi = param['pi']
    a = np.linspace(a_l, a_u, NA)
    util = np.full((NA, NA, NH), -10000.0)
    for ia in range(NA):
        for ih in range(NH):
            for iap in range(NA):
                cons = w*h[ih] + (1.0 + r)*a[ia] + T - a[iap]
                if cons > 0:
                    util[iap, ia, ih] = cons**(1.0-sigma)/(1.0-sigma)
    v = np.zeros((NA, NH))
    c = np.zeros((NA, NH))
    aplus = np.zeros((NA, NH))
    v_new = np.zeros((NA, NH))
    iaplus = np.full((NA, NH), -10000)
    tol = 1e-6
    test = 10.0
    reward = np.zeros((NA, NA, NH))
    while test > tol:
        for ia in range(NA):
            for ih in range(NH):
                reward[:, ia, ih] = util[:, ia, ih]
                for ihp in range(NH):
                    reward[:, ia, ih] += beta*pi[ih, ihp]*v[:, ihp]
                v_new[ia, ih] = np.max(reward[:, ia, ih])
                iaplus[ia, ih] = np.argmax(reward[:, ia, ih])
        test = np.max(abs(v_new - v))
        v = v_new
    aplus = a[iaplus]
    for ia in range(NA):
        for ih in range(NH):
            c[ia, ih] = w*h[ih] + (1.0 + r)*a[ia] + T - aplus[ia, ih]
    return aplus, iaplus, c

T = 1
decisions_with_subsidy = solve_household_with_subsidy(param, r, w, T)
aplus_with_subsidy = decisions_with_subsidy[0]
c_with_subsidy = decisions_with_subsidy[2]

# 貯蓄率を計算してプロット
plt.figure(figsize=(10, 6))
plt.plot(a, aplus[:, 0] / (c[:, 0] + aplus[:, 0]), label='低生産性 - 基本ケース', color='blue')
plt.plot(a, aplus[:, 1] / (c[:, 1] + aplus[:, 1]), label='高生産性 - 基本ケース', color='red')
plt.plot(a, aplus_with_subsidy[:, 0] / (c_with_subsidy[:, 0] + aplus_with_subsidy[:, 0]), label='低生産性 - 一括補助金', linestyle='--', color='blue')
plt.plot(a, aplus_with_subsidy[:, 1] / (c_with_subsidy[:, 1] + aplus_with_subsidy[:, 1]), label='高生産性 - 一括補助金', linestyle='--', color='red')
plt.xlabel("現在の資産水準 ($a_t$)")
plt.ylabel("貯蓄率 ($\\frac{a_{t+1}}{c_t + a_{t+1}}$)")
plt.title('一括補助金導入前後の貯蓄率の比較')
plt.legend()
plt.grid(True)
plt.show()
```
